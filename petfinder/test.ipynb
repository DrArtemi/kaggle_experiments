```python
train_data = pd.read_csv("/kaggle/input/petfinder-pawpularity-score/train.csv")
test_data = pd.read_csv("/kaggle/input/petfinder-pawpularity-score/test.csv")
```

```python
from sklearn.model_selection import train_test_split

y = train_data.Pawpularity
X = train_data.drop(["Id", "Pawpularity"], axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=50,
    learning_rate=0.05
)
```

```python
from sklearn.metrics import mean_squared_error

model.fit(X_train, y_train)

valid_preds = model.predict(X_valid)
mean_squared_error(y_valid, valid_preds)
```

```python
X_test = test_data.drop("Id", axis=1)

test_preds = model.predict(X_test)
```

```python
output = pd.DataFrame({"Id": test_data.Id, "Pawpularity": np.around(test_preds, decimals=2)})
output.to_csv("submission.csv", index=False)
```
